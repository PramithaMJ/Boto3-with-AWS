## Pricing

```python
import boto3
# Note that you have to specify a region, otherwise boto3 will raise an exception
# You can find a list of regions here: https://docs.aws.amazon.com/AmazonRDS/latest/UserGuide/Concepts.RegionsAndAvailabilityZones.html
client = boto3.client("pricing", region_name="us-east-1")  
```

To obtatin the pricing of an AWS service, you at first need to identify the service code of the service.
You can find all service codes here: https://docs.aws.amazon.com/de_de/general/latest/gr/aws-service-information.html

You might need to add the **AWSPriceListServiceFullAccess** policy to your group

```python
test = client.get_products(ServiceCode='AmazonS3')
```

As you can see, the results are not very readable and extremely detailed. Luckily we are able to filter it:


```python
test
```

To grab all the available keyword to filter for, we can use the **describe_services(ServiceCode)** function

```python
client.describe_services(ServiceCode="AmazonS3")
```

```python
client.get_products(
    ServiceCode='AmazonS3',
    Filters=[
        {'Type': 'TERM_MATCH', 'Field': 'regionCode', 'Value': 'us-east-1'},
        {'Type': 'TERM_MATCH', 'Field': 'storageClass', 'Value': 'us-east-1'},

    ]
)
```
